```python
import pandas as pd # importa a biblioteca pandas
```

```python
df = pd.read_csv('imigrantes_canada.csv') # importa a base de dados
```

```python
df.head() # exibe as 4 primeiras linhas do dataframe
```

```python
df.info() # exibe as informações do dataframe
```

TRATAMENTO DE DADOS

```python
df.set_index(['País'], inplace=True) #Seta o index com a coluna país
anos = list(map(str,range(1980,2013)))  # converte os anos em string armazenados em uma lista
```

```python
anos # exibe como string
```

```python
brasil = df.loc['Brasil',anos] # pega os dados somente para o país
```

```python
brasil
```

```python
# cria um dicionário com os dados do Brasil
brasil_dic={
    'ano':brasil.index.tolist(), # transforma a coluna com o index em uma lista
    'imigrantes': brasil.values.tolist()} #transforma a coluna com os valores em uma lista
```

```python
df_brasil = pd.DataFrame(brasil_dic) #Criando dataframe
```

```python
df_brasil # exibe os valores
```

```python
df_brasil.info() # exibe informações sobre o novo dataframe criado
```

```python
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes']) #plota o gráfico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14) # ajuste de valores para o eixo x
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14) # ajuste de valores para o eixo y
plt.xlabel("Anos",fontsize=14) # titulo eixo x
plt.title('Imigração do Brasil para o Canadá',fontsize =18) # adiciona título no gráfico
plt.ylabel("Número de imigrantes",fontsize=14) # titulo eixo y
plt.show() #exibe a figura
```

```python
argentina = df.loc['Argentina',anos] # armazena os dados da argentina
df_argentina = pd.DataFrame({'ano':argentina.index.tolist(),
                             'imigrantes': argentina.values.tolist()})
```

```python
df_argentina
```

```python
plt.plot(df_argentina['ano'],df_argentina['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.xlabel('Anos',fontsize=14)
plt.ylabel('Número de imigrantes',fontsize=14)
plt.title('Imigração da Argentina para o Canadá')
plt.show()
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.plot(df_argentina['ano'],df_argentina['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.xlabel('Anos',fontsize=14)
plt.ylabel('Número de imigrantes',fontsize=14)
plt.title('Imigração da Argentina para o Canadá')
plt.show()
```

```python
df_comparacao = df.loc[['Brasil','Argentina'],anos] # pega duas colunas Brasil e Argentina
```

```python
df_comparacao = df_comparacao.T # matriz transposta
```

```python
df_comparacao.head()
```

```python
df_comparacao['Total']=df_comparacao['Brasil']+ df_comparacao['Argentina'] # soma duas colunas do dataframe
```

```python
plt.figure(figsize=(10,10)) # tamanho da figura
plt.plot(df_comparacao['Brasil'],label = 'Brasil',color ='blue')
plt.plot(df_comparacao['Argentina'],label= 'Argentina',color='red')
plt.plot(df_comparacao['Total'],label= 'Total',color='black')
plt.xlabel('Anos',fontsize=14)
plt.ylabel('Número de imigrantes',fontsize=14)
plt.legend(loc='upper left')
plt.title("Comparação de imigrantes do Brasil para o Canadá \n e da Argentina para o Canadá")
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([500,1000,1500,2000,2500,3000,3500,4000],fontsize=14)
plt.show()
```

```python
fig, ax  = plt.subplots(figsize=(8,4))  # criando uma figura
# ax - axes 
ax.plot(df_brasil['ano'],df_brasil['imigrantes'],lw=3,color = 'green',label='Brasil') # comando para plotar o grafico
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # altera o intervalo do eixo x
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
ax.set_title('Imigração do Brasil para o Canadá  \n 1980 a 2013',loc='left')
ax.set_xlabel('Ano',fontsize=14)
ax.set_ylabel('Nº de imigrantes',fontsize=14)
ax.legend()
plt.show()
```

```python
# Criando a figura
fig, axs = plt.subplots(1, 2, figsize=(15,5))

# Subplot 1: Gráfico de linhas
axs[0].plot(df_brasil['ano'], df_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')

axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
# Subplot 2: Boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Box-plot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()
```
